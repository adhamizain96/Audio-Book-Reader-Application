```python
#PyPDF2 - PDF toolkit that allows pdf manipulation in memory
#what does PyPDF2 do? - extracting, splitting, merging, cropping, encrypting and decrypting PDF files
import PyPDF2
#pyttsx3 - convert text-to-speech offline
import pyttsx3

#rb - open/read a binary file
pdfReader = PyPDF2.PdfFileReader(open('book.pdf', 'rb'))
pages = pdfReader.numPages
speaker = pyttsx3.init()

for num in range(6, pages):
    page = pdfReader.getPage(num)
    text = page.extractText()
    speaker.say(text)
    #speaker.runAndWait() - blocks while processing all currently queued commands
    speaker.runAndWait()
```
